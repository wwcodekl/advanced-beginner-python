### Composition

Objects can be composed from other objects - relationship ...  

```python
class Cookie(object):
    def __init__(self, scarf):
        self.scarf = scarf

class Scarf(object):
    def __init__(self, color):
        self.color = color

green_scarf = Scarf(color='green')
c1 = Cookie(scarf=green_scarf)
```

### Inheritance

Why do we need to inherit codes? Look at the codes below, what did you notice?

```python
class Employee(object):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        
    def full_name(self):
        return "{} {}".format(
            self.first_name, self.last_name)
            
class Customer(object):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        
    def full_name(self):
        return "{} {}".format(
            self.first_name, self.last_name)
```

### Multiple Inheritance

If we were to draw the relationships between `Employee`, `Customer` and `Person` classes above, how will it look like?

How about example codes below?  

```python
class Vehicle(object):
    def __init__(self, name, wheel):
        self.name = name
        self.wheel = wheel

    def move(self):
        print("I can't move")

class GroundVehicle(Vehicle):
    def move(self):
        print("I'm moving through the ground")

class FlyingVehicle(Vehicle):

    def __init__(self, name, wheel, wings):
        Vehicle.__init__(self, name, wheel) # super().__init__(name, wheel)
        self.wings = wings
        
    def move(self):
        print("I'm moving through the air")

class Airplane(FlyingVehicle, GroundVehicle):
    
    def __init__(self, name, wheel, wings, lavatory):
        FlyingVehicle.__init__(self, name, wheel, wings) # super().__init__(name, wheel, wings)
        self.lavatory = lavatory
    
```
